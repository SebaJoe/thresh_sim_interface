```python
def normalize_edits(document):
    flattened = {}
    flattened['source'] = document['source']
    flattened['target'] = document['target']
    flattened['additions'] = []
    
    for edit in document['edits']:
        category = edit['category']

        try:
            annotation = edit['annotation']
        except KeyError:
            continue
        
        if category == 'hol_eval':
            flattened['hol_eval'] = annotation.get('hol_question_1', {'val': None})['val']
        elif category == 'final_q':
            flattened['final_q'] = annotation.get('final_commands')
        elif category == 'pico_pop':
            flattened['pico_pop'] = {
                'answer': annotation.get('pico_pop_anno', {'val': None})['val'],
                'rationale': annotation.get('pico_pop_rat')
            }
        elif category == 'pico_inter':
            flattened['pico_inter'] = {
                'answer': annotation.get('pico_pop_inter', {'val': None})['val'],
                'rationale': annotation.get('pico_inter_rat')
            }
        elif category == 'pico_comp':
            flattened['pico_comp'] = {
                'answer': annotation.get('pico_pop_comp', {'val': None})['val'],
                'rationale': annotation.get('pico_comp_rat')
            }
        elif category == 'pico_out':
            flattened['pico_out'] = {
                'exhaustive': annotation.get('pico_out_exhaustive'),
                'answer': annotation.get('pico_pop_out', {'val': None})['val'],
                'rationale': annotation.get('pico_out_rat')
            }
        elif category == 'add_info':
            start, end = edit['output_idx'][0]
            txt = document['target'][start:end]

            add_annotation = edit.get('annotation')
            if not add_annotation:
                rationale = None
            else:
                rationale = add_annotation['hal_rationale']
           
            flattened['additions'].append({
                'text': txt,
                'factual': '????',
                'rationale': rationale
            })

    return flattened

pprint(normalize_edits(doc))
```

```python
def load_annotations(json_path):
    all = []
    with open(json_path) as fin:
        docs = json.load(fin)
        for doc in docs:
            all.append(normalize_edits(doc))
    return all
```

```python
annotator_a = load_annotations('hannah.json')
annotator_b = load_annotations('monika.json')
```

```python
question_labels = {
    'hol_eval': 'Holistic',
    'final_q': 'Commands',
    'pico_pop': 'Population',
    'pico_inter': 'Intervention',
    'pico_comp': 'Comparator',
    'pico_out': 'Outcome',
}

answer_labels = {
    'q4_1': '1 - accurate',
    'q4_2': '2 - vague',
    'q4_3': '3 - missing critical',
    'q4_4': '4 - absent',
    'q4_5': 'N/A'
}

holistic_labels = {
    "hol_child_1": "1 - not useful at all",
    "hol_child_2": "2 - in the middle",
    "hol_child_3": "3 - very useful"
}
```

```python
def doc_to_excel(ws, ann_a, ann_b):
    header_values = ['Source', 'Target', 'Question', 'Hannah', 'Feedback', 'Monika', 'Feedback']
    ws.append(header_values)
    
    ws['A2'] = ann_a['source']
    ws['B2'] = ann_a['target']
    
    i = 2
    q = 'hol_eval'
    ws[f'C{i}'] = labels[q]
    ws[f'D{i}'] = holistic_labels[ann_a[q]]
    ws[f'F{i}'] = holistic_labels[ann_b[q]]
    i += 1
    
    q = 'final_q'
    ws[f'C{i}'] = labels[q]
    ws[f'D{i}'] = ann_a.get(q, '???')
    ws[f'F{i}'] = ann_b.get(q, '???')
    i += 1
    
    AGREE = PatternFill(start_color="CEEDD0", end_color="CEEDD0", fill_type="solid")
    DISAGGRE = PatternFill(start_color="FCEAA6", end_color="FCEAA6", fill_type="solid")
    
    for q in ['pico_pop', 'pico_inter', 'pico_comp', 'pico_out']:
        answer_a = ann_a[q]['answer']
        answer_b = ann_b[q]['answer']
        
        ws[f'C{i}'] = labels[q]
        ws[f'D{i}'] = answer_labels.get(answer_a, '???')
        ws[f'F{i}'] = answer_labels.get(answer_b, '???')
    
        color = AGREE if answer_a == answer_b else DISAGGRE
        ws[f'D{i}'].fill = color
        ws[f'F{i}'].fill = color
        i += 1
    
        ws[f'C{i}'] = 'Rationale'
        ws[f'D{i}'] = ann_a[q]['rationale']
        ws[f'F{i}'] = ann_b[q]['rationale']
        i += 1

    for add_a, add_b in itertools.zip_longest(ann_a['additions'], ann_b['additions']):
        for column_letter in range(ord('C'), ord('G') + 1):
            cell = ws[chr(column_letter) + str(i)]
            cell.fill = PatternFill(start_color='DDDDDDDD', end_color='DDDDDDDD', fill_type='solid')

        
        ws[f'C{i}'] = 'Addition'
        ws[f'C{i+1}'] = 'Factual'
        ws[f'C{i+2}'] = 'Rationale'
        if add_a:
            ws[f'D{i}'] = add_a['text']
            ws[f'D{i+1}'] = add_a['factual']
            ws[f'D{i+2}'] = add_a['rationale']

        if add_b:
            ws[f'F{i}'] = add_b['text']
            ws[f'F{i+1}'] = add_b['factual']
            ws[f'F{i+2}'] = add_b['rationale']
        i += 3
        
    
    
    ws.column_dimensions['A'].width = 35 
    ws.column_dimensions['B'].width = 35 
    ws.column_dimensions['C'].width = 15
    ws.column_dimensions['D'].width = 30 
    ws.column_dimensions['E'].width = 30 
    ws.column_dimensions['F'].width = 30 
    ws.column_dimensions['G'].width = 30 
    
    ws.merge_cells('A2:A100')
    ws.merge_cells('B2:B100')
    
    for row in ws.iter_rows():
        for cell in row:
            cell.alignment = Alignment(wrap_text=True, vertical='top' )
    
    bold_font = Font(bold=True)
    gray_fill = PatternFill(start_color="CCCCCC", end_color="CCCCCC", fill_type="solid")
    
    for cell in ws[1]:
        cell.font = bold_font
        cell.fill = gray_fill
```

```python
from openpyxl.styles import Alignment, Font, PatternFill
from openpyxl import Workbook
wb = Workbook()

ix_completed = 17
for i, (ann_a, ann_b) in enumerate(zip(annotator_a[:ix_completed], annotator_b[:ix_completed])):
    ws = wb.create_sheet(f'HIT {i+1}')
    ws.sheet_view.zoomScale = 135
    doc_to_excel(ws, ann_a, ann_b)
del wb['Sheet'] # drop default sheet

wb.save("split2-review.xlsx")
```
